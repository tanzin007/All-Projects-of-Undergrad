```python
# ==============================================================================
# 🎯 7th-8th Semester CSE Portfolio Project: Custom 53M UNet Diffusion Model
# 🚀 Phase 2: Kaggle Hardware Migration (Automated TPU / GPU / CPU Selector)
# 📊 Training Range: Epoch 71 to 300 (Continuous Learning Pipeline)
# ==============================================================================

import os
import re
import math
import pickle
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

# ------------------------------------------------------------------------------
# 🖥️ 1. AUTOMATED HARDWARE ACCELERATOR SELECTOR (TPU vs GPU vs CPU)
# ------------------------------------------------------------------------------
print("⚙️ Detecting available hardware accelerator...")

# Check for Kaggle TPU Environment first (PyTorch XLA)
try:
    import torch_xla.core.xla_model as xm
    # Verify if TPU is actually assigned to the notebook
    _ = xm.xla_device()
    USE_TPU = True
    device = xm.xla_device()
    print("🖥️ System Alert: kaggle TPU detected! Optimizing pipeline for Tensor Processing Units.")
except (ImportError, RuntimeError):
    USE_TPU = False
    # Fallback to GPU or standard CPU
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"🖥️ System Alert: TPU not available. Falling back to: {device.type.upper()}")

# ------------------------------------------------------------------------------
# 📦 2. KAGGLE DATASET PATH CONFIGURATION & DATALOADER
# ------------------------------------------------------------------------------
IMAGE_DIR = "/kaggle/input/datasets/tanlikesmath/the-oxfordiiit-pet-dataset/images"
# Verification for sub-folder layers inside Kaggle's input system
if os.path.exists(os.path.join(IMAGE_DIR, "images")):
    IMAGE_DIR = os.path.join(IMAGE_DIR, "images")

print(f"📂 Scanning images inside target source: {IMAGE_DIR}")

class OxfordPetDiffusionDataset(Dataset):
    def __init__(self, img_dir, transform=None):
        self.img_dir = img_dir
        self.transform = transform
        all_files = os.listdir(img_dir)
        self.valid_images = []
        self.prompts = []
        self.unique_classes = set()
        
        for file in all_files:
            if file.endswith('.jpg') and not file.startswith('.'):
                breed_raw = re.sub(r'_\d+$', '', file.split('.')[0])
                breed_name = breed_raw.replace('_', ' ').title()
                
                # Oxford dataset capitalization rule: Capitalized first letter = CAT
                if file[0].isupper():
                    prompt = f"a photo of a {breed_name} cat"
                else:
                    prompt = f"a photo of a {breed_name} dog"
                
                self.valid_images.append(file)
                self.prompts.append(prompt)
                self.unique_classes.add(prompt)
                
        self.unique_classes = sorted(list(self.unique_classes))
        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.unique_classes)}
        print(f"🔍 Dataset Summary: Found {len(self.valid_images)} images spanning {len(self.unique_classes)} classes.")

    def __len__(self):
        return len(self.valid_images)

    def __getitem__(self, idx):
        img_name = os.path.join(self.img_dir, self.valid_images[idx])
        image = Image.open(img_name).convert("RGB")
        if self.transform:
            image = self.transform(image)
        prompt = self.prompts[idx]
        class_idx = self.class_to_idx[prompt]
        return image, torch.tensor(class_idx, dtype=torch.long)

transform_pipeline = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

dataset = OxfordPetDiffusionDataset(img_dir=IMAGE_DIR, transform=transform_pipeline)
NUM_CLASSES = len(dataset.unique_classes)
train_loader = DataLoader(dataset, batch_size=64, shuffle=True, drop_last=True)

# ------------------------------------------------------------------------------
# 🧱 3. 53M UNET MODEL ARCHITECTURE & NOISE SCHEDULER
# ------------------------------------------------------------------------------
class DDPMNoiseScheduler:
    def __init__(self, timesteps=1000, beta_start=1e-4, beta_end=0.02):
        self.timesteps = timesteps
        self.betas = torch.linspace(beta_start, beta_end, timesteps).to(device)
        self.alphas = 1.0 - self.betas
        self.alphas_cumprod = torch.cumprod(self.alphas, dim=0)
        
    def add_noise(self, x_start, t):
        noise = torch.randn_like(x_start)
        alpha_cumprod_t = self.alphas_cumprod[t].view(-1, 1, 1, 1)
        x_noisy = torch.sqrt(alpha_cumprod_t) * x_start + torch.sqrt(1 - alpha_cumprod_t) * noise
        return x_noisy, noise

noise_scheduler = DDPMNoiseScheduler()

class TimeClassEmbedding(nn.Module):
    def __init__(self, dim, num_classes):
        super().__init__()
        self.dim = dim
        self.time_mlp = nn.Sequential(nn.Linear(dim, dim * 4), nn.GELU(), nn.Linear(dim * 4, dim * 4))
        self.class_emb = nn.Embedding(num_classes, dim * 4)

    def forward(self, time, class_idx):
        half_dim = self.dim // 2
        embeddings = math.log(10000) / (half_dim - 1)
        embeddings = torch.exp(torch.arange(half_dim, device=device) * -embeddings)
        embeddings = time[:, None] * embeddings[None, :]
        embeddings = torch.cat((embeddings.sin(), embeddings.cos()), dim=-1)
        return self.time_mlp(embeddings) + self.class_emb(class_idx)

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, emb_dim):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(out_channels)
        self.emb_proj = nn.Linear(emb_dim, out_channels)
        self.shortcut = nn.Sequential()
        if in_channels != out_channels:
            self.shortcut = nn.Sequential(nn.Conv2d(in_channels, out_channels, kernel_size=1), nn.BatchNorm2d(out_channels))

    def forward(self, x, emb):
        h = F.gelu(self.bn1(self.conv1(x)))
        h = h + self.emb_proj(F.gelu(emb))[:, :, None, None]
        return F.gelu(self.bn2(self.conv2(h)) + self.shortcut(x))

class Custom20MUNet(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.init_conv = nn.Conv2d(3, 128, kernel_size=3, padding=1)
        self.emb_net = TimeClassEmbedding(dim=128, num_classes=num_classes)
        emb_dim = 128 * 4

        self.down1 = nn.ModuleList([ResidualBlock(128, 128, emb_dim), ResidualBlock(128, 128, emb_dim)])
        self.down2 = nn.ModuleList([ResidualBlock(128, 256, emb_dim), ResidualBlock(256, 256, emb_dim)])
        self.down3 = nn.ModuleList([ResidualBlock(256, 512, emb_dim), ResidualBlock(512, 512, emb_dim)])
        self.down4 = nn.ModuleList([ResidualBlock(512, 512, emb_dim), ResidualBlock(512, 512, emb_dim)])

        self.mid1 = ResidualBlock(512, 512, emb_dim)
        self.mid_attention = nn.MultiheadAttention(embed_dim=512, num_heads=4, batch_first=True)
        self.mid2 = ResidualBlock(512, 512, emb_dim)

        self.up4 = nn.ModuleList([ResidualBlock(1024, 512, emb_dim), ResidualBlock(512, 512, emb_dim)])
        self.up3 = nn.ModuleList([ResidualBlock(1024, 256, emb_dim), ResidualBlock(256, 256, emb_dim)])
        self.up2 = nn.ModuleList([ResidualBlock(512, 128, emb_dim), ResidualBlock(128, 128, emb_dim)])
        self.up1 = nn.ModuleList([ResidualBlock(256, 128, emb_dim), ResidualBlock(128, 128, emb_dim)])
        self.final_conv = nn.Conv2d(128, 3, kernel_size=3, padding=1)

    def forward(self, x, t, class_idx):
        emb = self.emb_net(t, class_idx)
        x1 = self.init_conv(x)
        h1 = x1
        for block in self.down1: h1 = block(h1, emb)
        h2 = F.max_pool2d(h1, 2)
        for block in self.down2: h2 = block(h2, emb)
        h3 = F.max_pool2d(h2, 2)
        for block in self.down3: h3 = block(h3, emb)
        h4 = F.max_pool2d(h3, 2)
        for block in self.down4: h4 = block(h4, emb)
        
        m = F.max_pool2d(h4, 2)
        m = self.mid1(m, emb)
        b, c, h, w = m.shape
        m_flat = m.view(b, c, -1).transpose(1, 2)
        attn_out, _ = self.mid_attention(m_flat, m_flat, m_flat)
        m = attn_out.transpose(1, 2).view(b, c, h, w)
        m = self.mid2(m, emb)
        
        u4 = F.interpolate(m, scale_factor=2, mode='bilinear', align_corners=True)
        u4 = torch.cat([u4, h4], dim=1)
        for block in self.up4: u4 = block(u4, emb)
        u3 = F.interpolate(u4, scale_factor=2, mode='bilinear', align_corners=True)
        u3 = torch.cat([u3, h3], dim=1)
        for block in self.up3: u3 = block(u3, emb)
        u2 = F.interpolate(u3, scale_factor=2, mode='bilinear', align_corners=True)
        u2 = torch.cat([u2, h2], dim=1)
        for block in self.up2: u2 = block(u2, emb)
        u1 = F.interpolate(u2, scale_factor=2, mode='bilinear', align_corners=True)
        u1 = torch.cat([u1, h1], dim=1)
        for block in self.up1: u1 = block(u1, emb)
        return self.final_conv(u1)

# Initialize Model Structure
model = Custom20MUNet(num_classes=NUM_CLASSES).to(device)
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"📊 Model Parameter Count: {total_params / 1e6:.2f} Million")

# ------------------------------------------------------------------------------
# 📥 4. LOAD THE 70-EPOCH PRE-TRAINED WEIGHTS FROM KAGGLE INPUT PATH (FIXED FOR TPU)
# ------------------------------------------------------------------------------
MODEL_WEIGHTS_PATH = "/kaggle/input/models/edmunddantez/custom-53m-unet-diffusion-model/pytorch/default/1/custom_53m_diffusion_pet_70epochs.pth"

if os.path.exists(MODEL_WEIGHTS_PATH):
    print(f"📦 Loading pre-trained 70-epoch checkpoint from: {MODEL_WEIGHTS_PATH}")
    
    # FIX: Force mapping to CPU first to prevent UntypedStorage xla error
    checkpoint_weights = torch.load(MODEL_WEIGHTS_PATH, map_location="cpu")
    
    # Load the weights into the model structure
    model.load_state_dict(checkpoint_weights)
    
    # Push the model with newly loaded weights directly to your TPU/XLA device
    model = model.to(device)
    
    print("✅ Model weights mapped via CPU and pushed to TPU successfully! Ready to resume training.")
else:
    print(f"⚠️ Warning: Checkpoint file not found at {MODEL_WEIGHTS_PATH}. Training will start from scratch.")

# ------------------------------------------------------------------------------
# ⚡ 5. RESUMED TRAINING LOOP SETUP (EPOCH 71 TO 300)
# ------------------------------------------------------------------------------
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4)
criterion = nn.MSELoss()

START_EPOCH = 70
TOTAL_TARGET_EPOCHS = 300
loss_history = []

print(f"\n🎬 Resuming pipeline! Executing Epoch {START_EPOCH + 1} to {TOTAL_TARGET_EPOCHS}...")
model.train()

for epoch in range(START_EPOCH, TOTAL_TARGET_EPOCHS):
    epoch_loss = 0.0
    progress_bar = tqdm(enumerate(train_loader), total=len(train_loader), desc=f"Epoch {epoch+1}/{TOTAL_TARGET_EPOCHS}")
    
    for batch_idx, (images, class_idxs) in progress_bar:
        images = images.to(device)
        class_idxs = class_idxs.to(device)
        
        # Generate random timesteps using batch size dimension
        t = torch.randint(0, noise_scheduler.timesteps, (images.shape[0],), device=device).long()
        x_noisy, noise = noise_scheduler.add_noise(images, t)
        
        optimizer.zero_grad()
        predicted_noise = model(x_noisy, t, class_idxs)
        
        loss = criterion(predicted_noise, noise)
        loss.backward()
        
        # TPU execution optimization optimizer step
        if USE_TPU:
            xm.optimizer_step(optimizer, barrier=True)
        else:
            optimizer.step()
            
        epoch_loss += loss.item()
        progress_bar.set_postfix({"Loss": f"{loss.item():.4f}", "Avg Loss": f"{epoch_loss / (batch_idx + 1):.4f}"})
            
    average_epoch_loss = epoch_loss / len(train_loader)
    loss_history.append(average_epoch_loss)
    print(f"📢 Epoch {epoch+1} Complete! Final Average MSE Loss: {average_epoch_loss:.4f}\n")

print("🏆 Complete Extended 300 Epoch Training Finished Successfully!")

# ------------------------------------------------------------------------------
# 💾 6. SAVE PROGRESS TO KAGGLE WORKING DIRECTORY (SAVING THE ENTIRE MODEL OBJ)
# ------------------------------------------------------------------------------
final_model_name = "custom_53m_diffusion_pet_300epochs.pth"

# Modified to save the entire model object architecture along with parameters
torch.save(model, final_model_name)
print(f"✅ Final 300-epoch ENTIRE MODEL object saved in working directory as: {final_model_name}")

# ------------------------------------------------------------------------------
# 📊 7. DISPLAY FINAL PROGRESS LOSS GRAPH
# ------------------------------------------------------------------------------
plt.figure(figsize=(10, 5))
plt.plot(range(START_EPOCH + 1, TOTAL_TARGET_EPOCHS + 1), loss_history, marker='o', linestyle='-', color='b', label='Kaggle Shift Loss')
plt.title('Diffusion Model: Kaggle Resumed Training History (71 to 300 Epochs)')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# ==============================================================================
# 🎨 Part 1.2 (FULLY STANDALONE): 300-Epoch Model Inference & Generation
# 🎯 Goal: Initialize dataset mapping, noise scheduler, model architecture, 
#          load the 300-epoch checkpoint, and generate clean pet images.
# ==============================================================================

import os
import re
import math
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset
import numpy as np
import matplotlib.pyplot as plt

# ⚙️ 1. EXPLICIT KAGGLE PATH CONFIGURATIONS
# This points directly to the Oxford Pet images directory in Kaggle
IMAGE_DIR = "/kaggle/input/datasets/tanlikesmath/the-oxfordiiit-pet-dataset/images"
if os.path.exists(os.path.join(IMAGE_DIR, "images")):
    IMAGE_DIR = os.path.join(IMAGE_DIR, "images")

# Path to your uploaded 300-epoch model weights
MODEL_PATH = "/kaggle/input/models/edmunddantez/custom-53m-unet-diffusion-model-300-epoch/pytorch/default/1/custom_53m_diffusion_pet_300epochs.pth"

# 🖥️ 2. DEVICE ACCELERATION SETUP
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🖥️ Inference Accelerator: {device.type.upper()}")

# 📂 3. INFERENCE DATASET CLASS (To rebuild class mapping strings)
class OxfordPetInferenceDataset(Dataset):
    def __init__(self, img_dir):
        all_files = os.listdir(img_dir)
        self.unique_classes = set()
        
        for file in all_files:
            if file.endswith('.jpg') and not file.startswith('.'):
                # Extract breed name as a clean string token
                breed_raw = re.sub(r'_\d+$', '', file.split('.')[0])
                breed_name = breed_raw.replace('_', ' ').title()
                
                # Maintain the strict uppercase=CAT and lowercase=DOG logic from training
                if file.isupper():
                    prompt = f"a photo of a {breed_name} cat"
                else:
                    prompt = f"a photo of a {breed_name} dog"
                self.unique_classes.add(prompt)
                
        self.unique_classes = sorted(list(self.unique_classes))
        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.unique_classes)}

print("📂 Indexing dataset classes from directory structure...")
dataset = OxfordPetInferenceDataset(img_dir=IMAGE_DIR)
NUM_CLASSES = len(dataset.unique_classes)
print(f"🔍 Successfully re-mapped {NUM_CLASSES} unique text prompts.")

# ⏳ 4. NOISE SCHEDULER INITIALIZATION
class DDPMNoiseScheduler:
    def __init__(self, timesteps=1000, beta_start=1e-4, beta_end=0.02):
        self.timesteps = timesteps
        self.betas = torch.linspace(beta_start, beta_end, timesteps).to(device)
        self.alphas = 1.0 - self.betas
        self.alphas_cumprod = torch.cumprod(self.alphas, dim=0)

noise_scheduler = DDPMNoiseScheduler()

# 🧱 5. 53M PARAMETER UNET ARCHITECTURE BLOCK (Resolves __main__.Custom20MUNet AttributeError)
class TimeClassEmbedding(nn.Module):
    def __init__(self, dim, num_classes):
        super().__init__()
        self.dim = dim
        self.time_mlp = nn.Sequential(nn.Linear(dim, dim * 4), nn.GELU(), nn.Linear(dim * 4, dim * 4))
        self.class_emb = nn.Embedding(num_classes, dim * 4)

    def forward(self, time, class_idx):
        half_dim = self.dim // 2
        embeddings = math.log(10000) / (half_dim - 1)
        embeddings = torch.exp(torch.arange(half_dim, device=device) * -embeddings)
        embeddings = time[:, None] * embeddings[None, :]
        embeddings = torch.cat((embeddings.sin(), embeddings.cos()), dim=-1)
        return self.time_mlp(embeddings) + self.class_emb(class_idx)

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, emb_dim):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(out_channels)
        self.emb_proj = nn.Linear(emb_dim, out_channels)
        self.shortcut = nn.Sequential()
        if in_channels != out_channels:
            self.shortcut = nn.Sequential(nn.Conv2d(in_channels, out_channels, kernel_size=1), nn.BatchNorm2d(out_channels))

    def forward(self, x, emb):
        h = F.gelu(self.bn1(self.conv1(x)))
        h = h + self.emb_proj(F.gelu(emb))[:, :, None, None]
        return F.gelu(self.bn2(self.conv2(h)) + self.shortcut(x))

class Custom20MUNet(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.init_conv = nn.Conv2d(3, 128, kernel_size=3, padding=1)
        self.emb_net = TimeClassEmbedding(dim=128, num_classes=num_classes)
        emb_dim = 128 * 4

        self.down1 = nn.ModuleList([ResidualBlock(128, 128, emb_dim), ResidualBlock(128, 128, emb_dim)])
        self.down2 = nn.ModuleList([ResidualBlock(128, 256, emb_dim), ResidualBlock(256, 256, emb_dim)])
        self.down3 = nn.ModuleList([ResidualBlock(256, 512, emb_dim), ResidualBlock(512, 512, emb_dim)])
        self.down4 = nn.ModuleList([ResidualBlock(512, 512, emb_dim), ResidualBlock(512, 512, emb_dim)])

        self.mid1 = ResidualBlock(512, 512, emb_dim)
        self.mid_attention = nn.MultiheadAttention(embed_dim=512, num_heads=4, batch_first=True)
        self.mid2 = ResidualBlock(512, 512, emb_dim)

        self.up4 = nn.ModuleList([ResidualBlock(1024, 512, emb_dim), ResidualBlock(512, 512, emb_dim)])
        self.up3 = nn.ModuleList([ResidualBlock(1024, 256, emb_dim), ResidualBlock(256, 256, emb_dim)])
        self.up2 = nn.ModuleList([ResidualBlock(512, 128, emb_dim), ResidualBlock(128, 128, emb_dim)])
        self.up1 = nn.ModuleList([ResidualBlock(256, 128, emb_dim), ResidualBlock(128, 128, emb_dim)])
        self.final_conv = nn.Conv2d(128, 3, kernel_size=3, padding=1)

    def forward(self, x, t, class_idx):
        emb = self.emb_net(t, class_idx)
        x1 = self.init_conv(x)
        h1 = x1
        for block in self.down1: h1 = block(h1, emb)
        h2 = F.max_pool2d(h1, 2)
        for block in self.down2: h2 = block(h2, emb)
        h3 = F.max_pool2d(h2, 2)
        for block in self.down3: h3 = block(h3, emb)
        h4 = F.max_pool2d(h3, 2)
        for block in self.down4: h4 = block(h4, emb)
        
        m = F.max_pool2d(h4, 2)
        m = self.mid1(m, emb)
        b, c, h, w = m.shape
        m_flat = m.view(b, c, -1).transpose(1, 2)
        attn_out, _ = self.mid_attention(m_flat, m_flat, m_flat)
        m = attn_out.transpose(1, 2).view(b, c, h, w)
        m = self.mid2(m, emb)
        
        u4 = F.interpolate(m, scale_factor=2, mode='bilinear', align_corners=True)
        u4 = torch.cat([u4, h4], dim=1)
        for block in self.up4: u4 = block(u4, emb)
        u3 = F.interpolate(u4, scale_factor=2, mode='bilinear', align_corners=True)
        u3 = torch.cat([u3, h3], dim=1)
        for block in self.up3: u3 = block(u3, emb)
        u2 = F.interpolate(u3, scale_factor=2, mode='bilinear', align_corners=True)
        u2 = torch.cat([u2, h2], dim=1)
        for block in self.up2: u2 = block(u2, emb)
        u1 = F.interpolate(u2, scale_factor=2, mode='bilinear', align_corners=True)
        u1 = torch.cat([u1, h1], dim=1)
        for block in self.up1: u1 = block(u1, emb)
        return self.final_conv(u1)

# 📦 6. UNPICKLE AND ALLOCATE CHECKPOINT TO GPU
if os.path.exists(MODEL_PATH):
    print(f"📦 Safely unpickling 300-epoch model object from: {MODEL_PATH}")
    model = torch.load(MODEL_PATH, map_location="cpu", weights_only=False)
    model = model.to(device)
    model.eval()
    print("✅ Full model object successfully unpickled into fresh CUDA memory!")
else:
    print(f"❌ Error: Model file not found at '{MODEL_PATH}'. Double-check your input dataset link.")

# 🔮 7. REVERSE DDPM GENERATION PIPELINE
@torch.no_grad()
def generate_pet_images(model, noise_scheduler, class_name, num_images=4):
    model.eval()
    if class_name not in dataset.class_to_idx:
        print(f"❌ Class '{class_name}' not found.")
        return

    class_idx = dataset.class_to_idx[class_name]
    class_idxs = torch.tensor([class_idx] * num_images, dtype=torch.long).to(device)
    print(f"🎨 Generating {num_images} images for prompt: '{class_name}'...")

    # Start from pure Gaussian Noise
    x = torch.randn((num_images, 3, 64, 64)).to(device)

    # Denoising loop (1000 steps back to 0)
    for t_step in reversed(range(0, noise_scheduler.timesteps)):
        t_batch = torch.full((num_images,), t_step, device=device, dtype=torch.long)
        predicted_noise = model(x, t_batch, class_idxs)

        alpha = noise_scheduler.alphas[t_step]
        alpha_cumprod = noise_scheduler.alphas_cumprod[t_step]
        beta = noise_scheduler.betas[t_step]

        noise = torch.randn_like(x) if t_step > 0 else torch.zeros_like(x)
        x_mean = (1 / torch.sqrt(alpha)) * (x - ((1 - alpha) / torch.sqrt(1 - alpha_cumprod)) * predicted_noise)
        x = x_mean + torch.sqrt(beta) * noise

    # Image normalization and permutation
    x = torch.clamp(x, -1.0, 1.0)
    x = (x + 1.0) / 2.0
    x = x.cpu().permute(0, 2, 3, 1).numpy()

    # Rendering results grid
    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))
    for i in range(num_images):
        axes[i].imshow(x[i])
        title_display = class_name.split()[-2] if len(class_name.split()) >= 2 else class_name
        axes[i].set_title(f"Gen: {title_display}")
        axes[i].axis('off')
    plt.tight_layout()
    plt.show()

# 🚀 8. INFERENCE SCRIPT TRIGGER
if 'model' in locals():
    # Automatically takes the first valid prompt mapped from the input folder
    sample_prompt = dataset.unique_classes[0]
    generate_pet_images(model, noise_scheduler, class_name=sample_prompt, num_images=4)
```

```python
# ==============================================================================
# 🎨 Part 1.2 (FULLY STANDALONE): 300-Epoch Model Inference & Generation
# 🎯 Goal: Initialize dataset mapping, noise scheduler, model architecture, 
#          load the 300-epoch checkpoint, and generate clean pet images.
# ==============================================================================

import os
import re
import math
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset
import numpy as np
import matplotlib.pyplot as plt

# ⚙️ 1. EXPLICIT KAGGLE PATH CONFIGURATIONS
IMAGE_DIR = "/kaggle/input/datasets/tanlikesmath/the-oxfordiiit-pet-dataset/images"
if os.path.exists(os.path.join(IMAGE_DIR, "images")):
    IMAGE_DIR = os.path.join(IMAGE_DIR, "images")

# Path to your uploaded 300-epoch model weights
MODEL_PATH = "/kaggle/input/models/edmunddantez/custom-53m-unet-diffusion-model-300-epoch/pytorch/default/1/custom_53m_diffusion_pet_300epochs.pth"

# 🖥️ 2. DEVICE ACCELERATION SETUP
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🖥️ Inference Accelerator: {device.type.upper()}")

# 📂 3. INFERENCE DATASET CLASS (To rebuild class mapping strings)
class OxfordPetInferenceDataset(Dataset):
    def __init__(self, img_dir):
        all_files = os.listdir(img_dir)
        self.unique_classes = set()
        
        for file in all_files:
            if file.endswith('.jpg') and not file.startswith('.'):
                # Extract breed name as a clean string token
                breed_raw = re.sub(r'_\d+$', '', file.split('.')[0])
                breed_name = breed_raw.replace('_', ' ').title()
                
                # Maintain the strict uppercase=CAT and lowercase=DOG logic from training
                if file.isupper():
                    prompt = f"a photo of a {breed_name} cat"
                else:
                    prompt = f"a photo of a {breed_name} dog"
                self.unique_classes.add(prompt)
                
        self.unique_classes = sorted(list(self.unique_classes))
        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.unique_classes)}

print("📂 Indexing dataset classes from directory structure...")
dataset = OxfordPetInferenceDataset(img_dir=IMAGE_DIR)
NUM_CLASSES = len(dataset.unique_classes)
print(f"🔍 Successfully re-mapped {NUM_CLASSES} unique text prompts.")

# ⏳ 4. NOISE SCHEDULER INITIALIZATION
class DDPMNoiseScheduler:
    def __init__(self, timesteps=1000, beta_start=1e-4, beta_end=0.02):
        self.timesteps = timesteps
        self.betas = torch.linspace(beta_start, beta_end, timesteps).to(device)
        self.alphas = 1.0 - self.betas
        self.alphas_cumprod = torch.cumprod(self.alphas, dim=0)

noise_scheduler = DDPMNoiseScheduler()

# 🧱 5. 53M PARAMETER UNET ARCHITECTURE BLOCK (Resolves __main__.Custom20MUNet AttributeError)
class TimeClassEmbedding(nn.Module):
    def __init__(self, dim, num_classes):
        super().__init__()
        self.dim = dim
        self.time_mlp = nn.Sequential(nn.Linear(dim, dim * 4), nn.GELU(), nn.Linear(dim * 4, dim * 4))
        self.class_emb = nn.Embedding(num_classes, dim * 4)

    def forward(self, time, class_idx):
        half_dim = self.dim // 2
        embeddings = math.log(10000) / (half_dim - 1)
        embeddings = torch.exp(torch.arange(half_dim, device=device) * -embeddings)
        embeddings = time[:, None] * embeddings[None, :]
        embeddings = torch.cat((embeddings.sin(), embeddings.cos()), dim=-1)
        return self.time_mlp(embeddings) + self.class_emb(class_idx)

class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, emb_dim):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(out_channels)
        self.emb_proj = nn.Linear(emb_dim, out_channels)
        self.shortcut = nn.Sequential()
        if in_channels != out_channels:
            self.shortcut = nn.Sequential(nn.Conv2d(in_channels, out_channels, kernel_size=1), nn.BatchNorm2d(out_channels))

    def forward(self, x, emb):
        h = F.gelu(self.bn1(self.conv1(x)))
        h = h + self.emb_proj(F.gelu(emb))[:, :, None, None]
        return F.gelu(self.bn2(self.conv2(h)) + self.shortcut(x))

class Custom20MUNet(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.init_conv = nn.Conv2d(3, 128, kernel_size=3, padding=1)
        self.emb_net = TimeClassEmbedding(dim=128, num_classes=num_classes)
        emb_dim = 128 * 4

        self.down1 = nn.ModuleList([ResidualBlock(128, 128, emb_dim), ResidualBlock(128, 128, emb_dim)])
        self.down2 = nn.ModuleList([ResidualBlock(128, 256, emb_dim), ResidualBlock(256, 256, emb_dim)])
        self.down3 = nn.ModuleList([ResidualBlock(256, 512, emb_dim), ResidualBlock(512, 512, emb_dim)])
        self.down4 = nn.ModuleList([ResidualBlock(512, 512, emb_dim), ResidualBlock(512, 512, emb_dim)])

        self.mid1 = ResidualBlock(512, 512, emb_dim)
        self.mid_attention = nn.MultiheadAttention(embed_dim=512, num_heads=4, batch_first=True)
        self.mid2 = ResidualBlock(512, 512, emb_dim)

        self.up4 = nn.ModuleList([ResidualBlock(1024, 512, emb_dim), ResidualBlock(512, 512, emb_dim)])
        self.up3 = nn.ModuleList([ResidualBlock(1024, 256, emb_dim), ResidualBlock(256, 256, emb_dim)])
        self.up2 = nn.ModuleList([ResidualBlock(512, 128, emb_dim), ResidualBlock(128, 128, emb_dim)])
        self.up1 = nn.ModuleList([ResidualBlock(256, 128, emb_dim), ResidualBlock(128, 128, emb_dim)])
        self.final_conv = nn.Conv2d(128, 3, kernel_size=3, padding=1)

    def forward(self, x, t, class_idx):
        emb = self.emb_net(t, class_idx)
        x1 = self.init_conv(x)
        h1 = x1
        for block in self.down1: h1 = block(h1, emb)
        h2 = F.max_pool2d(h1, 2)
        for block in self.down2: h2 = block(h2, emb)
        h3 = F.max_pool2d(h2, 2)
        for block in self.down3: h3 = block(h3, emb)
        h4 = F.max_pool2d(h3, 2)
        for block in self.down4: h4 = block(h4, emb)
        
        m = F.max_pool2d(h4, 2)
        m = self.mid1(m, emb)
        b, c, h, w = m.shape
        m_flat = m.view(b, c, -1).transpose(1, 2)
        attn_out, _ = self.mid_attention(m_flat, m_flat, m_flat)
        m = attn_out.transpose(1, 2).view(b, c, h, w)
        m = self.mid2(m, emb)
        
        u4 = F.interpolate(m, scale_factor=2, mode='bilinear', align_corners=True)
        u4 = torch.cat([u4, h4], dim=1)
        for block in self.up4: u4 = block(u4, emb)
        u3 = F.interpolate(u4, scale_factor=2, mode='bilinear', align_corners=True)
        u3 = torch.cat([u3, h3], dim=1)
        for block in self.up3: u3 = block(u3, emb)
        u2 = F.interpolate(u3, scale_factor=2, mode='bilinear', align_corners=True)
        u2 = torch.cat([u2, h2], dim=1)
        for block in self.up2: u2 = block(u2, emb)
        u1 = F.interpolate(u2, scale_factor=2, mode='bilinear', align_corners=True)
        u1 = torch.cat([u1, h1], dim=1)
        for block in self.up1: u1 = block(u1, emb)
        return self.final_conv(u1)

# 📦 6. UNPICKLE AND ALLOCATE CHECKPOINT TO GPU
if os.path.exists(MODEL_PATH):
    print(f"📦 Safely unpickling 300-epoch model object from: {MODEL_PATH}")
    model = torch.load(MODEL_PATH, map_location="cpu", weights_only=False)
    model = model.to(device)
    model.eval()
    print("✅ Full model object successfully unpickled into fresh CUDA memory!")
else:
    print(f"❌ Error: Model file not found at '{MODEL_PATH}'. Double-check your input dataset link.")

# 🔮 7. REVERSE DDPM GENERATION PIPELINE
@torch.no_grad()
def generate_pet_images(model, noise_scheduler, class_name, num_images=4):
    model.eval()
    if class_name not in dataset.class_to_idx:
        print(f"❌ Class '{class_name}' not found. Available classes: {list(dataset.class_to_idx.keys())[:5]}...")
        return

    class_idx = dataset.class_to_idx[class_name]
    class_idxs = torch.tensor([class_idx] * num_images, dtype=torch.long).to(device)

    print(f"🎨 Generating {num_images} images for prompt: '{class_name}'...")
    x = torch.randn((num_images, 3, 64, 64)).to(device)

    for t_step in reversed(range(0, noise_scheduler.timesteps)):
        t_batch = torch.full((num_images,), t_step, device=device, dtype=torch.long)
        predicted_noise = model(x, t_batch, class_idxs)

        alpha = noise_scheduler.alphas[t_step]
        alpha_cumprod = noise_scheduler.alphas_cumprod[t_step]
        beta = noise_scheduler.betas[t_step]

        noise = torch.randn_like(x) if t_step > 0 else torch.zeros_like(x)
        x_mean = (1 / torch.sqrt(alpha)) * (x - ((1 - alpha) / torch.sqrt(1 - alpha_cumprod)) * predicted_noise)
        x = x_mean + torch.sqrt(beta) * noise

    x = torch.clamp(x, -1.0, 1.0)
    x = (x + 1.0) / 2.0
    x = x.cpu().permute(0, 2, 3, 1).numpy()

    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))
    for i in range(num_images):
        axes[i].imshow(x[i])
        title_display = class_name.split()[-2] if len(class_name.split()) >= 2 else class_name
        axes[i].set_title(f"Gen: {title_display}")
        axes[i].axis('off')
    plt.tight_layout()
    plt.show()

# 🚀 8. INFERENCE SCRIPT TRIGGER
if 'model' in locals():
    sample_prompt = dataset.unique_classes[0]
    generate_pet_images(model, noise_scheduler, class_name=sample_prompt, num_images=4)
```
